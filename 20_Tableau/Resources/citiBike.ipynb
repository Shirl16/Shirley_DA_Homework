```python
import pandas as pd
import datetime
import numpy as np
import os
```

```python
filepath1 = os.path.join("JC-201901-citibike-tripdata.csv")
filepath2 = os.path.join("JC-201902-citibike-tripdata.csv")
filepath3 = os.path.join("JC-201903-citibike-tripdata.csv")

jan2019 = pd.read_csv(filepath1)
jan2019.head()
```

```python
feb2019 = pd.read_csv(filepath2)
feb2019.head()
```

```python
mar2019 = pd.read_csv(filepath3)
mar2019.head()
```

```python
allBikeRides = pd.concat([jan2019, feb2019, mar2019])
allBikeRides.count()
```

```python
allBikeRides_df = allBikeRides
allBikeRides_df.head()
```

```python
# convert gender column values to M, F, unknown
allBikeRides_df['gender'] = allBikeRides_df['gender'].replace(
    {1: 'Male', 2: 'Female', 0: 'Unknown'})

allBikeRides_df.head()
```

```python
# convert trip duration from seconds to minutes
allBikeRides_df['tripduration'] = allBikeRides_df['tripduration']/60
allBikeRides_df.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
allBikeRides_df.to_csv("allBikeRides.csv", index=False, header=True)
```
