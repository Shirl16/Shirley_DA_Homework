```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# File to Load (Remember to change these)
city_data_to_load = "./data/city_data.csv"
ride_data_to_load = "./data/ride_data.csv"

# Read the City and Ride Data
# city = city, driver count, type
cityData_df = pd.read_csv("./data/city_data.csv")

# ride = city, date, fare, ride_id
rideData_df = pd.read_csv("./data/ride_data.csv")
#rideData_df.head()

# Combine the data into a single dataset

pyberData_df = pd.merge(rideData_df, cityData_df, how="left", on="city")


# Display the data table for preview
pyberData_df.head()
```

```python
# group by city
groupByCity = pyberData_df.groupby("city")

groupByCity.head(5)

# cityDriversOnly = cityData_df['driver_count'].count()
# cityDriversOnly
```

```python
# categorize each city by type
cityTypes_df = groupByCity['type'].first()
#cityTypes_df.head()

# # count drivers in each city= len 120
cityDrivers_df = groupByCity['driver_count'].first()
#cityDrivers_df.head()

# # calculate the average fare in each city
avgCityFare_df = groupByCity['fare'].mean()
#avgCityFare_df.head()

# # count rides in each city = len 120
ridesPerCity_df = groupByCity['ride_id'].count()
#ridesPerCity_df.head()
```

```python
# isolate urban cities in a dataframe
urbanCities_df = pyberData_df.loc[pyberData_df['type'] == "Urban"]
urbanCityGrouped = urbanCities_df.groupby('city')
#urbanCityGrouped.head()

urbanAvgFare = urbanCityGrouped['fare'].mean()
#urbanAvgFare


urbanDrivers = urbanCityGrouped['driver_count'].count()
#urbanDrivers

# plt.scatter(urbanDrivers,urbanAvgFare, s=urbanDrivers*10, alpha=0.75, color='red')
# plt.show()

```

```python
# isolate rural city data into a dataframe
ruralCities_df = pyberData_df.loc[pyberData_df['type'] == "Rural"]
ruralCitiesGrouped = ruralCities_df.groupby('city')

ruralAvgFare = ruralCitiesGrouped['fare'].mean()

ruralDrivers = ruralCitiesGrouped['driver_count'].count()


# plt.scatter(ruralDrivers,ruralAvgFare, s=ruralDrivers*10, alpha=1, color='yellow')
# plt.show()


```

```python
# isolate suburban cities into a dataframe
suburbanCities_df = pyberData_df.loc[pyberData_df['type'] == "Suburban"]
suburbanCitiesGrouped = suburbanCities_df.groupby('city')

suburbanAvgFare = suburbanCitiesGrouped['fare'].mean()

suburbanDrivers = suburbanCitiesGrouped['driver_count'].count()

# plt.scatter(suburbanDrivers,suburbanAvgFare, s=suburbanDrivers*10, alpha=1, color='blue')

# plt.show()


```

## Bubble Plot of Ride Sharing Data

```python
### Obtain the x and y coordinates for each of the three city types

x = ridesPerCity_df
y = avgCityFare_df


### Build the scatter plots for each city types

plt.scatter(suburbanDrivers,suburbanAvgFare, s=suburbanDrivers*11, marker='o', edgecolors='black',
           alpha=0.75, label='Suburban')

plt.scatter(ruralDrivers,ruralAvgFare, s=ruralDrivers*11, marker='o', edgecolors='black',
           alpha=0.75, label='Rural')

plt.scatter(urbanDrivers,urbanAvgFare, s=urbanDrivers*11, marker='o', edgecolors='black',
           alpha=0.75, label='Urban')


### Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid()


### Create a legend

plt.legend(loc='best', fancybox=True, title="City Types")

# Incorporate a text label regarding circle size- couldn't figure this out




### Save Figure
plt.savefig("./images/pyber_scatter_plot.png")
```

## Total Fares by City Type

```python
# # Calculate Type Percents
totalNumberFares = pyberData_df['fare'].sum()

urbanFares = urbanCities_df['fare'].sum()
urbanFarePercent = (urbanFares / totalNumberFares) * 100

ruralFares = ruralCities_df['fare'].sum()
ruralFaresPercent = (ruralFares / totalNumberFares) * 100

suburbanFares = suburbanCities_df['fare'].sum()
suburbanFaresPercent = (suburbanFares / totalNumberFares) * 100

labels = ("Urban", "Rural", "Suburban")
percentsByType = (urbanFarePercent, ruralFaresPercent, suburbanFaresPercent)

# Build Pie Chart
plt.title("% of Total Fares by City Type")
explode = (0.1, 0, 0)
plt.pie(percentsByType, labels=labels, explode=explode,autopct="%1.1f%%",
        shadow=True, startangle=290)
plt.axis("equal")

# Save Figure
plt.savefig("./images/pyber_fare_pie_plot.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
totalNumberRides = pyberData_df['ride_id'].count()
totalNumberRides

urbanRides = urbanCities_df['ride_id'].count()
urbanRidePercent = (urbanRides / totalNumberRides) * 100

ruralRides = ruralCities_df['ride_id'].count()
ruralRidesPercent = (ruralRides / totalNumberRides) * 100

suburbanRides = suburbanCities_df['ride_id'].count()
suburbanRidesPercent = (suburbanRides / totalNumberRides) * 100

# Build Pie Chart
labels = ("Urban", "Rural", "Suburban")
percentRidesByType = (urbanRidePercent, ruralRidesPercent, suburbanRidesPercent)

plt.title("% of Total Rides by City Type")
explode = (0.1, 0, 0)
plt.pie(percentRidesByType, labels=labels, explode=explode, autopct="%1.1f%%",
        shadow=True, startangle=260)
plt.axis("equal")


# Save Figure
plt.savefig("./images/pyber_rides_pie_plot.png")
```

```python
# Show Figure
plt.show()
```
