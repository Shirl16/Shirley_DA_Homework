## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# topSpenders 

topSpenders = purchase_data_df.groupby("SN", as_index=True).agg({"Item ID": "count"})
#topSpenders

topSpendersPrice = purchase_data_df.groupby("SN", as_index=True).agg({"Price": "mean"})
#topSpendersPrice

topSpendersTotalRevenue = purchase_data_df.groupby("SN", as_index=True).agg({"Price": "sum"})
#topSpendersTotalRevenue.head()
```

## Most Popular Items

```python
topSpenders_df = pd.DataFrame(topSpenders)
topSpenders_df = topSpenders_df.rename(columns={"Item ID":"Purchase Count"})
topSpenders_df.sort_values("Purchase Count", ascending=False)

topSpendersPrice_df = pd.DataFrame(topSpendersPrice)
topSpendersPrice_df = topSpendersPrice_df.rename(columns={"Price":"Average Purchase Price"})
topSpendersPrice_df["Average Purchase Price"] = topSpendersPrice_df[
    "Average Purchase Price"].map("${:,.2f}".format)
topSpendersPrice_df.sort_values("Average Purchase Price", ascending=False)

topSpendersTotalRevenue_df = pd.DataFrame(topSpendersTotalRevenue)
topSpendersTotalRevenue_df = topSpendersTotalRevenue_df.rename(columns={"Price": "Total Purchase Revenue"})
topSpendersTotalRevenue_df["Total Purchase Revenue"] = topSpendersTotalRevenue_df[
    "Total Purchase Revenue"].map("${:,.2f}".format)
topSpendersTotalRevenue_df.sort_values("Total Purchase Revenue", ascending=False)

mergeSpenders_df = pd.merge(topSpenders_df, topSpendersPrice_df, on="SN")
mergeSpenders_df

mergeSpendersRevenue_df = pd.merge(mergeSpenders_df, topSpendersTotalRevenue_df, on="SN")
mergeSpendersRevenue_df.sort_values("Total Purchase Revenue", ascending=False).head(5)
```

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting

```python
popularItems_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
#popularItems_indexed = popularItems.set_index(["Item ID", "Item Name"])
popularItems_df


popularCount = popularItems_df.groupby("Item Name", as_index=True).agg({"Item Name": "nunique"})
popularCount_df = popularCount.rename(columns={"Item Name": "Purchase Count"})
popularCount_df

popularRevenue = popularItems_df.groupby("Item Name", as_index=True).agg({"Price": "sum"})
popularRevenue_df = popularRevenue.rename(columns={"Price": "Total Purchase Value"})
popularRevenue_df


mergeTable = pd.merge(popularItems_df, popularCount_df, on="Item Name")
mergeTable

mergeSummary = pd.merge(mergeTable, popularRevenue_df, on="Item Name")
#mergeSummaryNames = mergeSummary.groupby("Item Name", as_index=True).agg({"Item Name": "count"})
#mergeSummaryNames
mergeSummary.head()

#TODO: why isn't the table unique values?
```
